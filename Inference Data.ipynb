```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Data Cleaning:
1. Including Main `Genre` Only (String Manip)
2. Include Main `Content Rating` Only
3. Remove `Last Updated`, `Current Ver`, `Android Ver`
4. Bin `Rating`, `Review`, `Size`
5. Numerical data for `Installs`
6. Remove duplicate observations
For further cleaning, https://www.kaggle.com/sabasiddiqi/google-play-store-apps-data-cleaning

```python
df = pd.read_csv('googleplaystore.csv')
df
```

```python
df["Genres"] = df["Genres"].str.split(";", 1).str[0]
df
```

```python
df["Content Rating"] = df["Content Rating"].str.split(" ", 1).str[0]
df["Content Rating"].unique()
```

```python
df = df.drop(["Last Updated", "Current Ver", "Android Ver"], axis=1)
```

```python
bins = [0, 1, 2, 3, 4, 5]
df["Rating"] = pd.cut(df['Rating'], bins)
df["Rating"]
```

```python
df["Size"].str.contains('k')
```

```python
df["Size"].str.contains('M')
```

```python
df["Size"].str.contains('Varies')
```

```python
df["Size"] = df["Size"].str.replace('k', 'e+3')
df["Size"] = df["Size"].str.replace('M', 'e+6')
df["Size"] = df["Size"].replace('Varies with device', np.nan)
df["Size"] = df["Size"].replace('1,000+', np.nan)
df["Size"]
```

```python
df.Size = pd.to_numeric(df.Size)
df.Size.unique()
```

```python
df = df[df.Reviews != "3.0M"]
df
```

```python
df.Reviews = pd.to_numeric(df.Reviews)
df.Reviews
```
